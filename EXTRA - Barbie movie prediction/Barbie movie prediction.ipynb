```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('budget-gross.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:, 1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
print(y)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x.reshape(-1, 1), y)
```

```python
barbie_budget = 151000000 #https://www.the-numbers.com/movie/Barbie-(2023)#tab=summary
barbie_budget_2013 = barbie_budget/1.36 #https://www.in2013dollars.com/us/inflation/2013?amount=1
prediction = regressor.predict([[barbie_budget_2013]])
actual = 1439851567
print('predicted ={0}, actual = 1,439,851,567'.format(int(prediction*1.36)))
print('difference = ', actual/prediction*1.36)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x.reshape(-1, 1)), color = 'blue')
plt.scatter(barbie_budget_2013, actual, color = 'green')
plt.title('Budget vs Gross Profits (normalized to 2013 $)')
plt.xlabel('Budget')
plt.ylabel('Gross Profits ')
plt.show()
```
